```python
climadrop = clima.drop(columns=['date'])
```

```python
sns.pairplot(climadrop, hue='weather', markers=['o', '>', '<', 'v','^'])
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(data=climadrop, x='temp_max', hue='weather', bins=20, kde=True)
plt.title("DistribuciÃ³n de Temperatura MÃ¡xima por CondiciÃ³n MeteorolÃ³gica")
plt.xlabel("Temperatura MÃ¡xima")
plt.ylabel("Frecuencia")
plt.show()
```

```python
# Matriz de correlaciÃ³n entre las variables numÃ©ricas
correlation_matrix = climadrop[['precipitation', 'temp_max', 'temp_min', 'wind']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title("Matriz de CorrelaciÃ³n")
plt.show()
```

```python
weather_counts = clima['weather'].value_counts()
sns.barplot(x=weather_counts.index, y=weather_counts.values)
plt.title('DistribuciÃ³n de Condiciones MeteorolÃ³gicas')
plt.xlabel('Condiciones MeteorolÃ³gicas')
plt.ylabel('Cantidad')
plt.xticks(rotation=45)
plt.show()
```
